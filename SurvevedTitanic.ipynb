```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
df = pd.read_csv('train.csv')
df.head(100)
```

```python
df = pd.read_csv('train.csv')

df['sex'] = df['sex'].map({'female': 1, 'male': 2})
# Xóa cột 'name'
df = df.drop(['name', 'home.dest'], axis=1)
# Thêm cột 'idpassenger' từ 0 đến cao
df['PassengerId'] = range(len(df))
df = df[['PassengerId'] + [col for col in df.columns if col != 'PassengerId']]
df['cabin'] = pd.factorize(df['cabin'])[0]
df['embarked'] = pd.factorize(df['embarked'])[0]
df['boat'] = pd.factorize(df['boat'])[0]
df['body'] = pd.factorize(df['body'])[0]
df['ticket'] = df['ticket'].str.extract('(\d+)', expand=False).astype(float).fillna(0).astype(int)
df = df.fillna(0)
df = df.reset_index(drop=True)
df.head(100)

# Chia dữ liệu thành features (X) và labels (y)
X = df.drop('survived', axis=1)  # Thay 'target_column_name' bằng tên cột chứa nhãn của bạn
y = df['survived']

# Chia dữ liệu thành tập huấn luyện và tập kiểm thử
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
X_train= X_train.values
X_test= X_test.values
y_train = y_train.values
y_test = y_test.values
df.head(11111)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score



i = 0
logreg_model = LogisticRegression(penalty='l1',  solver = 'liblinear', C=0.8, max_iter=10000, random_state=1)


logreg_model.fit(X_train, y_train)
predictions = logreg_model.predict(X_test)

for i in range(len(X_test)):
    print(f"Prediction: {predictions[i]}, Actual: {y_test[i]}")

accuracy = accuracy_score(predictions, y_test)
print("Accuracy:", accuracy)

y_pred= logreg_model.predict(X_test)
precision = precision_score(y_test, y_pred, average=None)
recall = recall_score(y_test, y_pred, average = None)
f1 = f1_score(y_test, y_pred, average= None)

for class_index in range(len(precision)):
	print(f'Class { class_index}:')
	print(f'Precision: {precision[class_index]}')
	print(f'Recall: {recall[class_index]}')
	print(f'F1-score: {f1[class_index]}')
for c_value in [0.00001, 0.01, 0.1, 1, 10, 100,10000]:
    logreg_model = LogisticRegression(penalty='l2', C=c_value, max_iter=1000, random_state=1)
    logreg_model.fit(X_train, y_train)
    predictions = logreg_model.predict(X_test)
    accuracy = accuracy_score(predictions, y_test)
    print(f"C = {c_value}, Accuracy = {accuracy}")
for penalty_type in ['l1', 'l2']:
    if penalty_type == 'l1':
        solver_type = 'liblinear'  # hoặc 'saga'
    else:
        solver_type = 'lbfgs'  # hoặc chọn solver khác hỗ trợ penalty 'l2'

    logreg_model = LogisticRegression(penalty=penalty_type, solver=solver_type, C=1.0, max_iter=1000, random_state=1)
    logreg_model.fit(X_train, y_train)
    predictions = logreg_model.predict(X_test)
    accuracy = accuracy_score(predictions, y_test)
    print(f"Penalty = {penalty_type}, Solver = {solver_type}, Accuracy = {accuracy}")
```
